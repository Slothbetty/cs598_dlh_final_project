```python
import pandas as pd
data=pd.read_csv(r'C:\Users\pengz3\Downloads\corpus_line.txt',header=None)
```

```python
data=data.rename(columns={0:'word'})
```

```python
words_counts=pd.value_counts(data.word)
```

```python
rare_words = words_counts[words_counts<5].index.tolist()
common_words = words_counts[words_counts>=5].index.tolist()
```

```python
len(rare_words)
```

```python
import json
with open(r'C:\Users\pengz3\Downloads\rare_common_dict_new.json') as json_file:
    rare_common_words_mapping= json.load(json_file)
```

```python
len(rare_common_words_mapping)
```

```python
from strsimpy.levenshtein import Levenshtein
import timeit
import simplejson as json
import datetime
import os
levenshtein = Levenshtein()
start = timeit.default_timer()
count = 0
for i in range(0,112608):
  count +=1
  if count %100 == 0:
    print('rare word count num: ', count)
    stop = timeit.default_timer()
    sec =  int(stop - start)
    time = str(datetime.timedelta(seconds = sec))
    print('Time: ', time)
  for c_word in common_words:
    current_distance = levenshtein.distance(rare_words[i], c_word)
    if rare_words[i] in rare_common_words_mapping:
      mapped_c_word = rare_common_words_mapping.get(rare_words[i])
      mapped_distance = levenshtein.distance(rare_words[i], mapped_c_word)
      if current_distance < mapped_distance:
        rare_common_words_mapping[rare_words[i]] = c_word
    else:
      rare_common_words_mapping[rare_words[i]] = c_word
  if count%1000 == 0:
    os.remove(r'C:\Users\pengz3\Downloads\rare_common_dict_old.json')
    os.rename(r'C:\Users\pengz3\Downloads\rare_common_dict_new.json',r'C:\Users\pengz3\Downloads\rare_common_dict_old.json')
    a_file = open(r'C:\Users\pengz3\Downloads\rare_common_dict_new.json', 'w')
    json.dump(rare_common_words_mapping, a_file)
    a_file.close()    
print(rare_common_words_mapping)
```
